Suele ser interesante representar las funciones para visualizar su comportamiento. Para ello utilizaremos la librería `matplotlib`.

```python
# Cargamos las librerías matemáticas
import numpy as np
#import sympy as sp

# Cargamos la librería para representación gráfica
import matplotlib.pyplot as plt
```

##### **Ejemplo**
Vamos a representar la función $f(x)=x \cos(x)$.

```python
# 1. Creamos una partición de un intervalo del eje x
x = np.arange(0,10,0.1) # -> "Partimos" el intervalo [0,10] en subintervalos de longitud 0.1

# 2. Calculamos la imagen de cada punto
y = x*np.cos(x)

# Pintamos el vector de puntos 'x' respecto al vector de imágenes 'y'
plt.plot(x,y)

# Añadimos etiquetas a los ejes
plt.xlabel('x')
plt.ylabel('y')

# Ponemos un título al gráfico
plt.title('Función $f(x)=x·cos(x)$')

# Mostramos el gráfico
plt.show()
```

##### ***Funciones***
Otra forma de trabajar con funciones en python es a partir de una definición "genérica" con la instrucción `def`.

```python
def funcion_ejemplo(x):
  return x+2
```

Lo bueno de definir así una función, es que podemos evaluar cualquier valor directamente, sin necesidad de la instrucción `.subs()`.

```python
funcion_ejemplo(3)
```

##### **Ejemplo**

Vamos a representar una función lineal y una función cuadrática en la misma gráfica, junto con los ejes coordenados.

```python
# Función lineal.
def f_lineal(x):
    return 4*x + 1

# Función cuadrática.
def f_cuadratica(x):
    return 2*(x**2) + 5*x - 2

# Valores del eje X que toma el gráfico.
x =  np.arange(-10,10,0.1)
  # NOTA: es interesante ver cómo cambia el gráfico si hacemos la partición más o menos fina
  # Es decir, si cambiamos el valor 0.1, por 1 o por 0.01.

# Graficamos ambas funciones
plt.plot(x, [f_lineal(i) for i in x])
plt.plot(x, [f_cuadratica(i) for i in x])

# Establecemos el color de los ejes
plt.axhline(0, color="black") # -> eje x negro
plt.axvline(0, color="red") # -> eje y rojo

# Limitamos los valores de los ejes.
plt.xlim(-10, 10)
plt.ylim(-10, 10)

# Podemos guardar el gráfico como una imagen PNG.
plt.savefig("grafico.png")

# Mostramos el gráfico
plt.show()
```
